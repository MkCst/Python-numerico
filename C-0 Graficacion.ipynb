````python
from random import randint
from pylab import plot, show, title, xlabel, ylabel, axis, legend


nombres=["maria","ricardo","fernando", "jose", "miguel"]
promedio_2018 = [randint(5,10) for _ in range(5)]
promedio_2019 = [randint(5,10) for _ in range(5)]
promedio_2020 = [randint(5,10) for _ in range(5)]

plot(nombres, promedio_2018)
plot(nombres, promedio_2019)
plot(nombres, promedio_2020)

#Customizacion de la grafica
legend(["promedio_2018", "promedio_2019", "promedio_2020"])
title("Promedio de alumnos")
xlabel("Nombres")
ylabel("Promedio")
axis(ymin=0)

#ajustando la grafica segun los valores


show()
````

#### Ley de gravitacion universal

Definida como:

$$ F=\frac{Gm_1m_2}{r^2}$$

Donde

$ G $ : es una constante con valor de $6.674x10^{-11}$

````python
# Se calcula la fuerza entre dos cuerpos en diferentes distancias
import matplotlib.pyplot as plt


def dibujar_grafico(x, y):
    plt.plot(x, y , marker="*")
    plt.xlabel("Distancia en metros")
    plt.ylabel("Fuerza gravitacional en Newtons")
    plt.title("Fuerza gravitacional y distancia")
    plt.show()

    
def generar_fuerza():
    #Genera valores de r
    r = range(100, 1001,50)
    
    # Variables involucradas
    m1 = 0.5
    m2 = 12.5
    
    # Constante
    G = 6.674 * (10**-11)
    
    #Lista para almacenar valores de F
    F=[G*(m1*m2)/(dist**2) for dist in r]
    
    dibujar_grafico(r, F)
    
generar_fuerza()
    
````

####  Movimiento de proyectil


````python
from matplotlib import pyplot as plt
import math

def frange(inicio, final, incremento):
    """floating point range"""
    numeros=[]
    while inicio < final:
        numeros.append(inicio)
        inicio+=incremento
    return numeros


def dibujar_grafico(x,y):
    plt.plot(x,y)
    plt.xlabel("Cordenada x")
    plt.ylabel("Cordenada y")
    plt.title("Movimiento del proyectil")
    
def dibujar_trayectoria(u, theta):
    theta = math.radians(theta)
    g = 9.8
    #Tiempo de vuelo
    t_vuelo = 2 * u * math.sin(theta) / g
    
    #Encuentra los intervalos de tiempo
    intervalos = frange(0, t_vuelo, 0.001)
    
    #Lista de las coordenadas x and y
    x=[]
    y=[]
    
    for t in intervalos:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t-0.5*g*t*t)
    
    dibujar_grafico(x,y)
    

#Programa principal
try:
    u = float(input("Ingrese la velocidad inicial m/s: "))
    theta = float(input("Ingrese el angulo de lanzamiento en grados: "))
    
except ValueError:
    print("Ingresaste un valor invalido")

else:
    dibujar_trayectoria(u, theta)
    plt.show()
````
